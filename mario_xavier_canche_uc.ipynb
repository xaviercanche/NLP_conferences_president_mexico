### **1. Cargue todas las conferencias en un string y aplique la función split para generar tokens fácilmente.**

```python
path_files = "../practica1/estenograficas_limpias_por_fecha2/"
```

```python
import glob

conferencias = ""

# Cargamos los archivos de la conferencia
for f_pagina in glob.glob(path_files+"/*"):
    # Leemos y concatenamos el contenido
    conferencias += open(f_pagina, "r", encoding="utf-8").read()

print("Tamaño de la cadena: ",len(conferencias))
```

```python
tokens = conferencias.split()
print("Total de tokens: ", len(tokens))
```

### **2. Contar la cantidad de palabras en todas las conferencias.**

```python
# Calculamos su longitud
total_tokens = len(tokens)
print("Total de tokens: ", total_tokens)
```

### **3. Extraer el vocabulario y mostrar su longitud.**

```python
# Extraemos el vocabulario
vocabulario = set(tokens)
```

```python
# Calculamos su longitud
total_vocabulario = len(vocabulario)
print("Tamaño del vocabulario: ", total_vocabulario)
```

### **4. Mida la riqueza del vocabulario de alguna forma en todos los documentos.**

```python
# Medimos la riqueza lexica
riqueza = total_tokens/total_vocabulario
print("La riqueza lexica para los documentos juntos es de: ", riqueza)
```

```python
# Procesamos de forma individual cada conferencia
for f_pagina in glob.glob(path_files+"/*"):
    # Leemos el contenido
    conferencia_individual = open(f_pagina, "r", encoding="utf-8").read()
    tokens = conferencia_individual.split()
    riqueza = len(tokens)/len(set(tokens))
    print("Riqueza de conferencia ",f_pagina, " es: ", riqueza)
```

### **6. Haga lo mismo que los puntos 1, 2 y 3 usando el tokenizador RegExp de NLTK, con una expresión regular que trate de sacar solo tokens que pueden contener solo letras mayúsculas y minúsculas. Después cargue los tokens en un objeto Text de NLTK en lugar de una lista de Python.**

```python
import nltk

# Expresión regular
pattern = r'\b[^\d\W]+\b'
```

```python
total_tokens_nltk = len(tokens_nltk)

print("Total de tokens cuando todo el texto esta en minusculas: ", total_tokens_nltk)
```

```python
# Extraemos el vocabulario
vocabulario_nltk = set(tokens_nltk)

# Calculamos su longitud
total_vocabulario_nltk = len(vocabulario_nltk)
print("Tamaño del vocabulario cuando todo el texto esta en minusculas: ", total_vocabulario_nltk)
```

```python
# Medimos la riqueza lexica
riqueza_nltk = total_tokens_nltk/total_vocabulario_nltk
print("La riqueza lexica cuando todo el texto esta en minusculas es de: ", riqueza_nltk)
```

```python
# Convertimos a objeto Text de NLTK
text = nltk.Text(tokens_nltk)
type(text)
```

### **5. Muestre un histograma de longitud en caracteres de las palabras. Muestre en el histograma el top 5 de longitudes más largas.**

```python
from nltk import FreqDist

# Calculamos la frecuencia de las longitudes de las palabras
fdist = FreqDist(len(w) for w in text_sort)
fdist
```

```python
# Histograma de las longitudes de las palabras
plt.figure(figsize=(20,5))
fdist.plot()
plt.show()
```

```python
# Tomamos las 5 longitudes mas largas
top5 = sorted(fdist.keys(), reverse=True)[:5]
top5
```

```python
# Calculamos su histograma en el corpus
fdist_top5 = FreqDist(len(w) for w in text_sort if len(w)>=top5[-1])
```

```python
fdist_top5
```

```python
# Graficamos el histograma
fdist_top5.plot()
plt.show()
```

## 4. Otras librerías en Python

### **1. Mencione dos librerías en Python además de NLTK para NLP. Ponga una desventaja y ventaja de cada una.**

**spaCy:** Otra librería popular para NLP similar a NLTK y que teóricamente puede realizar cualquier tarea de PNL. Una desventaja de esta librería sería que funciona como un tipo de servicio y no puedes contruir nada desde cero como harías en NLTK, lo que impide hacer desarrollo de investigación. Por el contrarió, esto resulta ser una ventaja para desarrolladores de aplicaciones ya que proporciona las versiones más rápidas y precisas de algoritmos de NLP existentes.

**Polyglot:** Otro paquete de Python para NLP. Tiene entre sus ventajas el ser eficiente, sencilla y rápida. Similar a Spacy, por lo que se puede usar en proyectos que involucren algún lenguaje que Spacy no admita. Funciona mejor en el procesamiento de texto/habla inusual o informal donde no se siguen las reglas del lenguaje natural. Entre sus desventajas podemos decir que es poco popular y se puede encontrar poca documentación o soporte comparado a las otras librerías como NLTK o spaCy.

### **2. Mencione tres alternativas para Text Processing en NLP qué existen en otros lenguajes. De una ventaja y desventaja de cada una.**

**Stanford CoreNLP:** Librería de NLP para JAVA. Una ventaja es que contiene wrappers para otros lenguajes, python incluido. Es más costoso computacionalmente comparado a otras librerias y no ofrece resultados significativamente superiores a estos.  

**Apache OpenNLP:** Otra Librería de NLP para JAVA. Entre sus ventajas esta que es fácil de usar y entender, mucha documentación y ejemplos, y cubre una gran cantidad de funciones para NLP. Entre sus desventajas podemos decir que a sus modelos pueden faltarles un mejor entrenamiento (dependiendo del caso de uso), algunos modelos no están y su desarrollo y actualización es muy lenta.  

**GATE:** (General Architecture for Text Engineering) es una arquitectura y una plataforma de desarrollo diseñada
en la Universidad de Sheffield para la elaboración e integración de aplicaciones en tecnologías del lenguaje. Se ejecuta en JAVA. Entre sus ventajas es que como comenzo a desarrollarse desde 1995, cuenta con un amplio abanico de funcionalidades de procesamiento de texto. Entre sus desventajas estaría que estas funcionalidades solo abarcan el idioma inglés y los otros idiomas solo cuentan con tokenizador y algunas funcionalidades básicas.
